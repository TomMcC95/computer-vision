```python
import cv2
import time
import matplotlib.pyplot as plt

#Initialize video capture.
cap = cv2.VideoCapture(1, cv2.CAP_DSHOW)

#Create blank frame array.
frame_array = []

#Mark start time
t = time.time()

#Loop until you hit the 'Esc' key or 10 seconds elapses.
while time.time() - t < 10:

    #Capture the current frame.
    ret, frame = cap.read()

    print(ret)

    #Append the image to frame array.
    frame_array.append(frame)

#Detect if the Esc key has been pressed.
    c = cv2.waitKey(1)
    if c == 27:
        break


# Release the video capture object
cap.release()
```

```python
import matplotlib.pyplot as plt
import cv2
import time

#Initialize video capture


def get_frames(number):

    cap = cv2.VideoCapture(1, cv2.CAP_DSHOW)

    frame_array = []
    t = time.time()
    
    #Loop until specific time elapses.
    while time.time() - t < (number/10):

        #Capture the current frame.
        ret, frame = cap.read()

        print(ret)

        #Append the image to frame array.
        frame_array.append(frame)
```
