```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('validation_1_distributions.xlsx')
sns.kdeplot(data=df1, x='Diameter', hue='Tool', common_norm=False)
val_1_mean_cal = df1[df1.Tool == 'Caliper']['Diameter'].mean()
val_1_mean_cam = df1[df1.Tool == 'Camera']['Diameter'].mean()
val_1_std_cal = df1[df1.Tool == 'Caliper']['Diameter'].std()
val_1_std_cam = df1[df1.Tool == 'Camera']['Diameter'].std()
val_1_mean_diff = val_1_mean_cal - val_1_mean_cam
val_1_std_diff = val_1_std_cal - val_1_std_cam
plt.text(0, 0.18, f'Std Dev Difference = {val_1_std_diff:.2f}\nMean Difference = {val_1_mean_diff:.2f}')
plt.title('Validation 1')
plt.savefig('Validation 1 Distribution.png', dpi=200)
```
